### 1. Missing Values

Question: Are there any missing values in the dataset, and if so, how should they be handled for each indicator?


```python
df.isnull().sum()
```

```python
# Replacing null values in Age column with Average Age

age_avg = round(df.Age.mean(),0)
age_avg
```

```python
df['Age'] = df.Age.fillna(age_avg)
```

```python
#Replaced null values in Rating column with zero and rating can't be negative so replaced -1 with 0

df['Rating'] = df['Rating'].fillna(0)
df['Rating'] = df['Rating'].replace(-1.0,0)
df['Rating']
```

## `3. Outliers`


Question: Identify potential outliers in numerical indicators (e.g., Age, Salary, Rating). Should outliers be removed or adjusted?


Established column's mean is very less and there's a lot of outliers in that column as there as many values as -1. It needs to be changed as year cannot be -1. I have decided to replace that value with mean after changing the -1 value to null.

```python
df[['Index','Age','Rating','Established']].describe()
```

```python
df['Established'] = df['Established'].replace(-1,np.NaN)
df['Established']
```

```python
#Replacing null with mean year which is 1980

df['Established'] = df['Established'].fillna(round(df['Established'].mean(),0))
```

```python
#Converting float to int

df['Established'] = df['Established'].apply(lambda x : int(x))
```

## 6. Established Column:
>Question: Explore the Established column. Are there any inconsistencies or anomalies that need to be addressed?


## 7. Easy Apply Indicator:

>Question: Analyze the Easy Apply column. Does it contain boolean values or need transformation for better analysis?

Easy Apply column contains Boolean values in string literal. So I converted it into boolean type



```python
df['Easy Apply'] = df['Easy Apply'].replace('TRUE',True)
df['Easy Apply'] = df['Easy Apply'].replace('FALSE',False)
df['Easy Apply'].dtype
```

### 8. Rating Range:
>Question: Investigate the range of values in the Rating column. Does it fall within expected rating scales, and how should outliers be treated?






```python
df['Rating'].describe()
```

Since I replaced the negative values with 0 the min, quartile and mean values were influenced by it so I replaced the zero with NaN value to find the mean. Assumption here is that the rating scale is between 0-10.

```python
new_df = df['Rating'].replace(0.0,np.NaN)
new_df.describe()
rating_mean = new_df.mean()
```

```python
df['Rating'] = df['Rating'].replace(0,rating_mean)
df['Rating'] = df['Rating'].apply(lambda x : round(x,1))

# After replacing with mean, the range of ratings is displayed
df['Rating'].describe()
```
