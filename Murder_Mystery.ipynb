```python
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect('sql-murder-mystery.db')
```

```python
print(conn)
```

```python
pd.set_option('display.max_colwidth', None)
```

```python
query_1 = '''
SELECT *
FROM crime_scene_report
WHERE city = 'SQL City'
ORDER BY date
'''

pd.read_sql_query(query_1,conn)
```

```python
# Checking the details of witnesses:
query_2 = '''
SELECT * FROM person
WHERE address_street_name = "Northwestern Dr"
ORDER BY address_number DESC
LIMIT 1;
'''
pd.read_sql_query(query_2,conn)
```

```python
query_3 = '''
SELECT * FROM person
WHERE address_street_name = "Franklin Ave"
AND name LIKE "Annabel%";
'''
pd.read_sql_query(query_3,conn)
```

```python
# Interview of Morty and Annabel (witnesses):
query_4 = '''
SELECT * FROM interview
WHERE person_id = 14887 OR person_id = 16371;
'''
pd.read_sql_query(query_4,conn)
```

```python
# Checking details from gym:
query_5 = '''
SELECT * FROM get_fit_now_member
WHERE id LIKE "48Z%" AND membership_status = "gold";
'''
pd.read_sql_query(query_5,conn)
```

```python
query_6 = '''
SELECT * FROM get_fit_now_check_in
WHERE check_in_date = 20180109;
'''
pd.read_sql_query(query_6,conn)
```

Both the persons with the membership starting with '48Z' have checked in on 9th Jan

```python
# Checking license plate:
query_7 = '''
SELECT * FROM drivers_license
WHERE plate_number LIKE "%H42W%";
'''
pd.read_sql_query(query_7,conn)
```

```python
# Checking personal details of suspects that we got from previous queries:
query_8 = '''
SELECT * FROM person
WHERE id IN (28819,67318);
'''
pd.read_sql_query(query_8,conn)
```

```python
query_9 = '''
SELECT * FROM person
WHERE license_id IN (183779,423327,664760);
'''
pd.read_sql_query(query_9,conn)
```

The person that matches both criteria with the gold gym membership that starts with '48Z' and license plate number that contains 'H42W' is 'JEREMY BOWERS'and he was also seen to be working out on JAN 9th. Hence **JEREMY BOWERS** is the murderer.
