## Adding the scraped information to a Pandas Dataframe

```python
import pandas as pd
```

```python
df = pd.DataFrame()
```

```python
x = {'title':title, 'rating':rating,'price':price,'link':book_link}

df=df.append(x,ignore_index = True)
```

```python
# Using for loop to get the information of 20 books in a page

for i in range(20):
    book = books_tag[i]

    title_tag = book.find('a',title=True)
    title = title_tag['title']

    rating = book.find('p')['class'][1]

    price = book.find('p',class_="price_color")
    price = price.text.strip('Â')

    link = book.find('a')['href']
    book_link = url+link

    x = {'title':title, 'rating':rating,'price':price,'link':book_link}

    df =df.append(x,ignore_index = True)
```

```python
df
```

```python
# Defining a function to get the info of each book that will be looped

def page(book):

    title_tag = book.find('a',title=True)
    title = title_tag['title']

    rating = book.find('p')['class'][1]

    price = book.find('p',class_="price_color")
    price = price.text.strip('Â')

    link = book.find('a')['href']
    book_link = url+link

    x = {'title':title, 'rating':rating,'price':price,'link':book_link}

    return x
```

```python
# Looping the url by changing the page number and getting to scrape 10 pages consisting of 200 books

url2 = 'https://books.toscrape.com/catalogue/page-2.html'
response = requests.get(url2)
soup2 = bs(response.text)
books_tag2 = soup2.find_all('article',class_='product_pod')
for i in range(20):
    book = books_tag2[i]
    temp = page(book)
    df = df.append(temp,ignore_index = True)

```

## Combining all the above steps into a single block.
## Creting a new df called books_df which has the information of books from 10 pages

```python
books_df  = pd.DataFrame(columns=['title','rating','price','link'])
books_df
```

```python
url = 'https://books.toscrape.com/catalogue/page-1.html'
for i in range(0,10):
  url1 = url.replace('1',str(i+1))
  response = requests.get(url1)
  soup = bs(response.text)
  books_tag = soup.find_all('article',class_='product_pod')
  for i in range(20):
      book = books_tag[i]
      temp = page(book)
      books_df = books_df.append(temp,ignore_index = True)
```

```python
books_df
```

As we can see the rows are all right aligned in the books_df. So making it all left aligned for uniformity and better readability

```python
left_aligned_df = books_df.style.set_properties(**{'text-align': 'left'})

left_aligned_df = left_aligned_df.set_table_styles(
    [dict(selector='th', props=[('text-align', 'left')])])
```
